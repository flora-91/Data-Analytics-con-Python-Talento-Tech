# Datos de ventas mensuales (en unidades)
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul',

---

* Primero importamos las librerías que vamos a usar: **Pandas** para manejar los datos y **Matplotlib** para graficar.
* Luego creamos dos listas con los nombres de los meses y las ventas mensuales, y las combinamos en un **DataFrame**, que es una estructura de datos tabular.
* Finalmente mostramos el contenido para verificar que todo esté correcto.


````python
# Importamos las librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt

# Creamos los datos del ejercicio
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul',
         'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
ventas = [100, 120, 130, 150, 180, 170, 200, 210, 250, 270, 300, 320]

# Creamos un DataFrame de Pandas para organizar los datos
df = pd.DataFrame({'Mes': meses, 'Ventas': ventas})

# Mostramos el DataFrame
df
````

### Cómo crear un gráfico de líneas en Matplotlib

El comando principal para crear un gráfico de líneas es `plt.plot()`.

Su sintaxis básica es:

```python
plt.plot(x, y)
```

Donde:

* `x` representa los valores del eje horizontal (por ejemplo, los meses).
* `y` representa los valores del eje vertical (por ejemplo, las ventas).

Un gráfico típico suele incluir además:

```python
plt.title("Título del gráfico")    # agrega un título
plt.xlabel("Etiqueta del eje X")   # etiqueta horizontal
plt.ylabel("Etiqueta del eje Y")   # etiqueta vertical
plt.grid(True)                     # agrega una cuadrícula
plt.show()                         # muestra el gráfico
```

En el siguiente bloque, usamos esta estructura para graficar las ventas mensuales.

````python
# Gráfico básico de líneas con Matplotlib
plt.plot(df["Mes"], df["Ventas"])  # definimos eje X y eje Y
plt.title("Ventas mensuales")      # título del gráfico
plt.xlabel("Mes")                  # etiqueta del eje X
plt.ylabel("Unidades vendidas")    # etiqueta del eje Y
plt.grid(True)                     # muestra la cuadrícula
plt.show()                         # muestra el gráfico en pantalla
````

Este es el gráfico más simple que podemos hacer con Matplotlib.

`plt.plot()` traza una línea que conecta los puntos (x, y). En este caso, los **meses** se ubican en el eje horizontal y las **ventas** en el vertical.

El resto de las líneas agrega el título, etiquetas y una cuadrícula para que sea más legible.

---

###  Mejora del gráfico de líneas

Ahora vamos a usar algunos parámetros de `plt.plot()` para hacerlo más claro y atractivo:

* `color` define el color de la línea (puede ser nombre, código RGB o hexadecimal).
* `linewidth` cambia el grosor.
* `marker` agrega un símbolo en cada punto de datos.
* `linestyle` define el tipo de trazo (continuo, punteado, etc.).
* En `plt.grid()` agregamos `alpha=0.5` para hacer la cuadrícula más tenue.


````python
# Gráfico de líneas mejorado con opciones de estilo

plt.plot(
    df["Mes"], df["Ventas"],
    color="royalblue",        # color de la línea
    linewidth=2.5,            # grosor
    marker="o",               # marcador de cada punto
    linestyle=":"             # estilo de línea: puede ser '-', '--', '-.', ':'
)

plt.title("Ventas mensuales (año anterior)")
plt.xlabel("Mes")
plt.ylabel("Unidades vendidas")
plt.grid(True, linestyle="--", alpha=0.5)  # cuadrícula punteada más suave
plt.show()
````

### Agregar etiquetas de valores a cada punto

Vamos a ver ahora cómo agregar **etiquetas numéricas** a cada punto del gráfico para que sea más informativo:

* El bucle `for` recorre cada valor de ventas y usa `plt.text()` para mostrar el número justo arriba de cada punto.

* `i` representa la posición del mes (porque los índices del DataFrame van de 0 a 11).

* `valor + 5` eleva un poco el texto para que no se superponga con el marcador.

* `ha="center"` centra la etiqueta sobre el punto.


````python
# Gráfico con etiquetas numéricas en cada punto

plt.plot(
    df["Mes"], df["Ventas"],
    color="royalblue",
    linewidth=2.5,
    marker="o"
)

# Recorremos los datos y colocamos una etiqueta sobre cada punto
for i, valor in enumerate(df["Ventas"]):
    plt.text(
        i, valor + 5,             # posición del texto (x, y)
        str(valor),               # texto mostrado
        ha="center", fontsize=9,  # alineación y tamaño
        color="darkblue"
    )

plt.title("Ventas mensuales (con valores)")
plt.xlabel("Mes")
plt.ylabel("Unidades vendidas")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
````

### Otras variaciones del gráfico de líneas

En el gráfico siguiente se superponen tres líneas con distintos estilos:

* `label=` permite identificar cada línea en la **leyenda** (activada con `plt.legend()`).
* Cambiamos el `marker` (círculo, cuadrado, triángulo).
* Usamos diferentes `linestyle`: `"--"` (guiones) y `":"` (puntos).
* `plt.figure(figsize=(10,5))` controla el tamaño del gráfico.


````python
# Variaciones del gráfico de líneas: distintos estilos y colores

plt.figure(figsize=(10, 5))  # definimos tamaño del gráfico

# Línea original
plt.plot(
    df["Mes"],
    df["Ventas"],
    color="royalblue",
    marker="o",
    label="Ventas")

# Línea con otro estilo
plt.plot(
    df["Mes"],
    df["Ventas"],
    color="orange",
    linestyle="--",
    marker="s",
    label="Tendencia (estimada)")

# Línea suavizada (simulamos proyección)
proyeccion = [v * 1.05 for v in df["Ventas"]]

plt.plot(
    df["Mes"],
    proyeccion,
    color="green",
    linestyle=":",
    marker="^",
    label="Proyección")

plt.title("Comparación de estilos con plt.plot()")
plt.xlabel("Mes")
plt.ylabel("Unidades vendidas")
plt.legend()   # agrega leyenda automática
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
````

###  Controlar la posición y estilo de la leyenda

* `loc` define la posición de la leyenda. Puede usarse una palabra clave (`'upper right'`, `'lower left'`, etc.) o un código numérico (por ejemplo, `loc=1` equivale a `'upper right'`).
* `fontsize` ajusta el tamaño del texto.
* `frameon` muestra o esconde el recuadro.
* `shadow` agrega una sombra al fondo.
* `facecolor` y `edgecolor` modifican los colores del recuadro.

> También se puede usar `loc="best"`, que deja que Matplotlib elija automáticamente la posición donde menos tapa los datos.



````python
# Ejemplo con leyenda personalizada

plt.figure(figsize=(10, 5))

plt.plot(
    df["Mes"],
    df["Ventas"],
    color="royalblue",
    marker="o",
    label="Ventas reales")

plt.plot(
    df["Mes"], [v * 1.05 for v in df["Ventas"]],
    color="green",
    linestyle="--",
    marker="s",
    label="Proyección")

# Personalización de la leyenda
plt.legend(
    loc="upper left",       # posición: 'upper left', 'upper right', 'lower left', etc.
    fontsize=10,            # tamaño del texto
    frameon=True,           # muestra el recuadro
    shadow=True,            # agrega sombra
    facecolor="whitesmoke", # color de fondo
    edgecolor="gray"        # color del borde
)

plt.title("Ejemplo de personalización de leyenda")
plt.xlabel("Mes")
plt.ylabel("Unidades vendidas")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
````

### Análisis e interpretación

El gráfico muestra un crecimiento sostenido de las ventas a lo largo del año.  

Se observa un aumento moderado durante los primeros meses y un crecimiento más marcado a partir de julio, alcanzando su punto máximo en diciembre.  

Este comportamiento puede indicar una **mayor demanda en la segunda mitad del año**, posiblemente relacionada con promociones o fechas especiales.

La tendencia general es positiva, lo que sugiere que la marca logró **incrementar sus ventas de forma constante** mes a mes.
