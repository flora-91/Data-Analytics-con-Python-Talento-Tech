* Finalmente mostramos el contenido para verificar que todo esté correcto.


````python
# Importamos las librerías
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Datos de ventas por producto
productos = ['Producto A', 'Producto B', 'Producto C', 'Producto D']
ventas = [300, 250, 400, 350]

# Creamos un DataFrame
df = pd.DataFrame({
    'Producto': productos,
    'Ventas': ventas
})

# Mostramos el DataFrame
df
````

### Gráfico de barras en Seaborn

Esta es la **versión equivalente con Seaborn**, usando un estilo moderno, colores agradables y un pequeño detalle para **resaltar el producto más vendido**:


### Diferencias y mejoras:

* Usamos `sns.barplot()` en lugar de `plt.bar()`: Seaborn maneja mejor los DataFrames y aplica automáticamente estilos agradables.
* Se añadió una **leyenda automática** para resaltar el producto más vendido.
* Colores coherentes con el tema moderno (`royalblue` y `lightgray`).
* Tamaño de figura y tipografía ajustados para presentación (`figsize`, `font_scale` desde antes).

````python
# Gráfico de barras simple
# Aplicamos un tema moderno y agradable
sns.set_theme(style="whitegrid", palette="deep", font_scale=1.1)

# Gráfico de barras con Seaborn

# Identificamos el producto con mayor venta
producto_top = df.loc[df["Ventas"].idxmax(), "Producto"]

# Creamos una columna auxiliar para destacar el más vendido
df["Destacado"] = df["Producto"].apply(lambda x: "Más vendido" if x == producto_top else "Otros")

# Creamos el gráfico
plt.figure(figsize=(8, 5))
sns.barplot(
    data=df,
    x="Producto",
    y="Ventas",
    hue="Destacado",        # diferenciamos con color
    palette={"Más vendido": "royalblue", "Otros": "lightgray"}
)

# Personalizamos el gráfico
plt.title("Ventas por producto", fontsize=14, weight="bold")
plt.xlabel("Producto")
plt.ylabel("Unidades vendidas")
plt.legend(title="", loc="upper left")
plt.grid(axis="y", linestyle="--", alpha=0.5)

# Mostramos
plt.show()

````

````python
# VERSION FINAL ( SEABORN)

# -----------------------------------------------------
# Configuración del estilo visual
# -----------------------------------------------------
sns.set_theme(style="whitegrid", palette="deep", font_scale=1.1)

# -----------------------------------------------------
# Dataset de ejemplo
# -----------------------------------------------------
productos = ['Producto A', 'Producto B', 'Producto C', 'Producto D']
ventas = [300, 250, 400, 350]

df = pd.DataFrame({
    "Producto": productos,
    "Ventas": ventas
})

# -----------------------------------------------------
# Identificar el producto más vendido y crear una columna auxiliar
# -----------------------------------------------------
producto_top = df.loc[df["Ventas"].idxmax(), "Producto"]   # obtenemos el nombre del producto con más ventas
df["Destacado"] = df["Producto"].apply(                    # nueva columna para colorear según categoría
    lambda x: "Más vendido" if x == producto_top else "Otros"
)

# -----------------------------------------------------
# Crear el gráfico
# -----------------------------------------------------
plt.figure(figsize=(8, 5))   # tamaño de la figura (ancho, alto)

sns.barplot(
    data=df,
    x="Producto",             # eje X: nombres de los productos
    y="Ventas",               # eje Y: unidades vendidas
    hue="Destacado",          # color según la categoría ("Más vendido" u "Otros")
    palette={                 # colores personalizados
        "Más vendido": "royalblue",
        "Otros": "lightgray"
    }
)

# -----------------------------------------------------
# Añadir etiquetas de valor encima de cada barra
# -----------------------------------------------------
for index, fila in df.iterrows():
    plt.text(
        x=index,                  # posición horizontal (coincide con el índice del producto)
        y=fila["Ventas"] + 5,     # un poco por encima de la barra
        s=f"{fila['Ventas']} u.", # texto a mostrar (unidades)
        ha='center',              # alineación horizontal centrada
        va='bottom',              # alineación vertical en la parte inferior del texto
        fontsize=10               # tamaño del texto
    )

# -----------------------------------------------------
# Personalización estética
# -----------------------------------------------------
plt.title("Ventas por producto", fontsize=14, weight="bold")  # título del gráfico
plt.xlabel("Producto")                                       # etiqueta del eje X
plt.ylabel("Unidades vendidas")                              # etiqueta del eje Y
plt.legend(title="", loc="upper left")                       # leyenda sin título, arriba a la izquierda
plt.grid(axis="y", linestyle="--", alpha=0.5)                # cuadrícula horizontal suave

# -----------------------------------------------------
# Mostrar el gráfico
# -----------------------------------------------------
plt.show()
````

### Y, por supuesto, la versión "horizontal":

````python
# VERSION HORIZONTAL (Seaborn)

# -----------------------------------------------------
# Configuración del estilo visual
# -----------------------------------------------------
sns.set_theme(style="whitegrid", palette="deep", font_scale=1.1)

# -----------------------------------------------------
# Dataset de ejemplo
# -----------------------------------------------------
productos = ['Producto A', 'Producto B', 'Producto C', 'Producto D']
ventas = [300, 250, 400, 350]

df = pd.DataFrame({
    "Producto": productos,
    "Ventas": ventas
})

# -----------------------------------------------------
# Identificar el producto más vendido y crear una columna auxiliar
# -----------------------------------------------------
producto_top = df.loc[df["Ventas"].idxmax(), "Producto"]   # obtenemos el nombre del producto con más ventas
df["Destacado"] = df["Producto"].apply(                    # nueva columna para colorear según categoría
    lambda x: "Más vendido" if x == producto_top else "Otros"
)

# -----------------------------------------------------
# Crear el gráfico de barras horizontal
# -----------------------------------------------------
plt.figure(figsize=(8, 5))   # tamaño de la figura (ancho, alto)

sns.barplot(
    data=df,
    y="Producto",             # eje Y: productos (ahora en el eje vertical)
    x="Ventas",               # eje X: unidades vendidas
    hue="Destacado",          # color según la categoría ("Más vendido" u "Otros")
    palette={                 # colores personalizados
        "Más vendido": "royalblue",
        "Otros": "lightgray"
    },
    dodge=False               # evita que las barras se separen verticalmente
)

# -----------------------------------------------------
# Añadir etiquetas con los valores al final de cada barra
# -----------------------------------------------------
for index, fila in df.iterrows():
    plt.text(
        x=fila["Ventas"] + 5,     # un poco a la derecha del final de la barra
        y=index,                  # posición vertical (coincide con el índice del producto)
        s=f"{fila['Ventas']} u.", # texto a mostrar (unidades)
        va='center',              # alineación vertical centrada
        fontsize=10               # tamaño del texto
    )

# -----------------------------------------------------
# Personalización estética
# -----------------------------------------------------
plt.title("Ventas por producto", fontsize=14, weight="bold")  # título del gráfico
plt.xlabel("Unidades vendidas")                              # etiqueta del eje X
plt.ylabel("Producto")                                       # etiqueta del eje Y
plt.legend(title="", loc="center left")                       # leyenda sin título
plt.grid(axis="x", linestyle="--", alpha=0.5)                # cuadrícula vertical suave

# -----------------------------------------------------
# Mostrar el gráfico
# -----------------------------------------------------
plt.show()
````
