✅ Paso 4: Parámetros de Palabras Clave Indefinidas (**kwargs)

**Consigna:** Crear una función llamada concatena_info que reciba un nombre y un número indefinido de palabras clave
(por ejemplo: edad, ciudad, ocupación) y devuelva un string que combine todos los datos.
Probar la función pasando varios pares de clave-valor.


```python
def concatena_info(nombre, **datos):
  mensaje = f"Nombre: {nombre}\n"
  for clave, valor in datos.items():
      mensaje += f"{clave}: {valor}\n"
  print(mensaje)


concatena_info("Juan", edad=10, colegio="ES1", altura=1.55)
```
