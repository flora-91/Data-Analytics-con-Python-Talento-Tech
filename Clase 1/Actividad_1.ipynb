Realizar un programa que pida la edad, la altura y el nombre.
Si es mayor de edad y mide mas de 160 cm, lo habilitamos a subirse a un juego de un parte.

```python
nombre = input("Nombre :")
edad = int(input("Edad....:"))
altura = int(input("Altura:"))

if edad >= 18:
  if altura >= 160:
    print(f"{nombre} bienvenido/a a la montaña rusa!")
  else:
    print("No pasas por la altura")
else:
  print("No pasas por la edad")
```
