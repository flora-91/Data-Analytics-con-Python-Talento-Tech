## Actividad 1: Análisis de Tendencia Central

Tenemos que calcular las **medidas de tendencia central** de las ventas mensuales de un producto específico en el último año. Esto nos ayudará
a identificar patrones y tomar decisiones informadas sobre el inventario.

### Objetivos:
* Calcular la media, la mediana y la moda de las ventas mensuales.
* Comprender la importancia de estas medidas en el análisis de datos de
ventas.


### Ejercicio práctico:
Realiza los siguientes cálculos en Python con los datos sobre las ventas mensuales en unidades del último año:
1. Calcular la media de las ventas.
2. Calcular la mediana de las ventas.
3. Calcular la moda de las ventas.

Interpretá y explicá qué significan estos valores.

```python
# Importamos Pandas:
import pandas as pd

# Este es nuestro set de datos:
ventas = {
  'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
          'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'],
  'Ventas (millones)': [1.2, 2.5, 3.1, 18.3, 40.5, 52.1,
                        54.8, 46.2, 25.5, 13.8, 11.9, 9.2]
}

# Convertimos a DataFrame y mostramos:
df = pd.DataFrame(ventas)
display(df)
```

### 1. Calcular la media de las ventas
La **media** es el **promedio**: indica el valor típico si las ventas se distribuyeran equitativamente entre todos los meses.

```python
media = df['Ventas (millones)'].mean()
print(f"Media de ventas: {media:.2f} millones")
```

### 2. Calcular la mediana de las ventas

La **mediana** es el valor central: separa a la mitad de los meses con ventas menores y la otra mitad con ventas mayores.


```python
mediana = df['Ventas (millones)'].median()
print(f"Mediana de ventas: {mediana:.2f} millones")
```

### 3. Calcular la moda de las ventas

La **moda** es el **valor que más se repite**.
Si no hay valores repetidos, no existe una moda definida.


```python
moda = df['Ventas (millones)'].mode()
display(moda)
```

### Conclusión:

* La **media** indica el promedio de ventas mensuales del año.
* La **mediana** refleja el punto medio de las ventas, útil para entender la tendencia general.
* La **moda**, al no existir en este caso, nos dice que todas las ventas fueron diferentes.

Estas tres medidas, analizadas juntas, nos permiten comprender mejor el comportamiento de las ventas: la **media** resume el rendimiento general, la **mediana** evita el efecto de los meses con valores extremos, y la **moda** ayuda a detectar patrones o repeticiones en los datos.


## Actividad 2: Medidas de Dispersión

Nos piden analizar la variabilidad de las ventas mensuales que calculamos en la actividad anterior

### Objetivos:
* Calcular el rango, la varianza y la desviación standard de las ventas
mensuales.
* Reconocer cómo estas medidas de dispersión informan sobre la estabilidad

1. Calcular el rango de las ventas.
2. Calcular la varianza de las ventas.

Interpretá y explicá qué significan estos valores.

### 1. Calcular el rango
El **rango** muestra la diferencia entre el mes con más ventas y el mes con menos ventas.

```python
rango = df['Ventas (millones)'].max() - df['Ventas (millones)'].min()
print(f"Rango: {rango:.2f} millones")
```

El rango es muy grande, lo que indica que hay una diferencia notable entre los meses de menor y mayor venta.
Esto sugiere un producto con **ventas muy estacionales**.

### 2. Calcular la varianza
La **varianza** mide cuánto se alejan, en promedio, los valores respecto de la media.

```python
varianza = df['Ventas (millones)'].var()
print(f"Varianza: {varianza:.2f}")
```

Una varianza alta implica que las ventas fluctúan mucho entre un mes y otro.
Si la varianza fuera pequeña, significaría que las ventas se mantienen bastante estables durante todo el año.


#### Nota importante:
¿Que significa "alta" en este contexto? La varianza está expresada en "millones al cuadrado" y por eso por sí sola no dice mucho. Puede parecer alta, pero no sabemos si lo es hasta tener un punto de referencia.


Hay tres formas comunes de saber si la variabilidad es "alta" o "baja":

#### a) Comparándola con la **media**

* Si es del mismo orden o mayor que la media, hay **mucha variabilidad**.
* Si es pequeña en comparación con la media, las ventas son **estables**.


#### b) Comparándola con otros productos o periodos
Si analizás ventas de varios productos, podés ver cuál tiene la varianza más alta:

* Alta varianza: ventas muy irregulares.
* Baja varianza: ventas más estables.


#### c) Mirando el gráfico o la distribución

Un simple gráfico de líneas o barras permite ver si los valores están concentrados o muy dispersos. Esto lo veremos la proxima clase...pero va un spoiler:

```python
df.plot(x='Mes', y='Ventas (millones)', kind='line', marker='o', grid=True, figsize=(6,4))
```

### 3. Calcular la desviación estándar
La **desviación estándar** indica, en las mismas unidades que los datos, la magnitud típica de esas variaciones.

```python
desviacion = df['Ventas (millones)'].std()
print(f"Desviación estándar: {desviacion:.2f} millones")
```

Una desviación estándar alta confirma que las ventas son muy inestables a lo largo del año.

### **Análisis final**

Las tres medidas muestran una gran dispersión: el rango es alto, la varianza y la desviación estándar también.
Esto indica que el producto tiene **picos fuertes en ciertos meses** y **ventas bajas el resto del año**.
