## Identificar filas duplicadas en un DataFrame

** Consigna:** Usar el método `duplicated()` para identificar las filas duplicadas y contar cuántas filas duplicadas hay en total.

**Tips:**

- `df.duplicated()` devuelve una serie de valores booleanos (`True` si la fila es duplicada, `False` si no lo es).
- `df.duplicated().sum()` cuenta cuántas filas duplicadas existen en total.
- `df[df.duplicated()]` muestra todas las filas duplicadas.



```python
# df.duplicated()
display(df.duplicated().sum())

# Sumamos los NULOS de cada columna , y creamos una fila
# con la cantidad de NULOS que tiene cada columna.
display(df.isnull().sum())

display(df[df.duplicated()])
```

# 3 - Informe Final

* 1. Cantidad total de registros en el DataFrame
* 2. Cantidad total de valores nulos por columna
* 3. Cantidad de filas duplicadas


```python
# Cantidad total de registros en el DataFrame
total = len(df)

# Cantidad total de valores nulos por columna
nulos = df.isnull().sum()

# Cantidad de filas duplicadas
duplicados = df.duplicated().sum()

# Imprimir el informe
print ("---------------------------------")
print ("            Informe")
print ("---------------------------------")
print ("Cantidad total de registros:", total)
print ("Cantidad total de valores nulos por columna:\n", nulos)
print ("Cantidad de filas duplicadas:",duplicados)
```

## 4 - Crear un DataFrame con las filas duplicadas para que se vea el contenido

Para ver el contenido de los registros duplicados:

- Usamos `df.duplicated()` para marcar qué filas son duplicadas.
- Filtramos el DataFrame original con esa condición.
- Guardamos el resultado en una nueva variable, por ejemplo `df_duplicados`.

Así obtenemos un DataFrame con solo los duplicados.


```python
df_dupli = df[df.duplicated()]

print("La cantidad de duplicados son:", len(df_dupli))
```
